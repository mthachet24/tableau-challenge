```python
import pandas as pd
import numpy as np
```

```python
csv_2019 = "2019 june trip data.csv"
csv_2020 = "2020 june trip data.csv"

bike_trip_2019 = pd.read_csv(csv_2019)
bike_trip_2020 = pd.read_csv(csv_2020)
```

```python
bike_2019_df = bike_trip_2019
bike_2019_df.head()
```

```python
bike_2020_df = bike_trip_2020
bike_2020_df.head()
```

```python
del bike_2019_df['usertype']
del bike_2020_df['usertype']
```

```python
bike_2019_df.head()
```

```python
bike_2020_df.head()
```

```python
bike_2019_df.dtypes
```

```python
bike_2020_df.dtypes
```

```python
bike_2019_df['gender'] = bike_2019_df['gender'].astype(str)
bike_2020_df['gender'] = bike_2020_df['gender'].astype(str)
```

```python
checking_gender_2019 = bike_2019_df['gender']
male_2019 = checking_gender_2019.replace('1', 'Male')
female_2019 = male_2019.replace('2', 'Female')
unknown_2019 = female_2019.replace('0', 'Unknown')
bike_2019_df['gender'] = unknown_2019
bike_2019_df
```

```python
checking_gender_2020 = bike_2020_df['gender']
male_2020 = checking_gender_2020.replace('1', 'Male')
female_2020 = male_2020.replace('2', 'Female')
unknown_2020 = female_2020.replace('0', 'Unknown')
bike_2020_df['gender'] = unknown_2020
bike_2020_df
```

```python
current_year = 2021

birth_year_2019 = bike_2019_df['birth year']
age = current_year - birth_year_2019
bike_2019_df['age'] = age

birth_year_2020 = bike_2020_df['birth year']
age = current_year - birth_year_2020
bike_2020_df['age'] = age
```

```python
bike_2019_df_organized = bike_2019_df[["bikeid",
                                       "gender",
                                       "birth year",
                                       "age",
                                       "tripduration",
                                       "start station id",
                                       "start station name",
                                       "start station latitude",
                                       "start station longitude",
                                       "starttime",
                                       "end station id",
                                       "end station name", 
                                       "end station latitude",
                                       "end station longitude",
                                       "stoptime"]]

bike_2020_df_organized = bike_2020_df[["bikeid",
                                       "gender",
                                       "birth year",
                                       "age",
                                       "tripduration",
                                       "start station id",
                                       "start station name",
                                       "start station latitude",
                                       "start station longitude",
                                       "starttime",
                                       "end station id",
                                       "end station name", 
                                       "end station latitude",
                                       "end station longitude",
                                       "stoptime"]]

bike_2019_df_organized.head()
```

```python
bike_2020_df_organized.head()
```

```python
bike_2019_renamed_df = bike_2019_df_organized.rename(columns={"bikeid":"Bike ID", 
                                                              "gender":"Gender",
                                                              "birth year":"Birth Year",
                                                              "age":"Age",
                                                              "tripduration":"Trip Duration",
                                                              "start station id":"Start Station ID",
                                                              "start station name":"Start Station Name",
                                                              "start station latitude":"Start Station Latitude",
                                                              "start station longitude":"Start Station Longitude",
                                                              "starttime":"Start Time",
                                                              "end station id":"End Station ID",
                                                              "end station name":"End Station Name",
                                                              "end station latitude":"End Station Latitude",
                                                              "end station longitude":"End Station Longitude",
                                                              "stoptime":"Stop Time"})

bike_2020_renamed_df = bike_2020_df_organized.rename(columns={"bikeid":"Bike ID", 
                                                              "gender":"Gender",
                                                              "birth year":"Birth Year",
                                                              "age":"Age",
                                                              "tripduration":"Trip Duration",
                                                              "start station id":"Start Station ID",
                                                              "start station name":"Start Station Name",
                                                              "start station latitude":"Start Station Latitude",
                                                              "start station longitude":"Start Station Longitude",
                                                              "starttime":"Start Time",
                                                              "end station id":"End Station ID",
                                                              "end station name":"End Station Name",
                                                              "end station latitude":"End Station Latitude",
                                                              "end station longitude":"End Station Longitude",
                                                              "stoptime":"Stop Time"})



bike_2019_renamed_df
```
