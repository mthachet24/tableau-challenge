```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
joining_january = os.path.join("january 2020 bike data.csv")
joining_february = os.path.join("february 2020 bike data.csv")
joining_march = os.path.join("march 2020 bike data.csv")
joining_april = os.path.join("april 2020 bike data.csv")
joining_may = os.path.join("may 2020 bike data.csv")
joining_june = os.path.join("june 2020 bike data.csv")
joining_july = os.path.join("july 2020 bike data.csv")
joining_august = os.path.join("august 2020 bike data.csv")
joining_september = os.path.join("september 2020 bike data.csv")
joining_october = os.path.join("october 2020 bike data.csv")
joining_november = os.path.join("november 2020 bike data.csv")
joining_december = os.path.join("december 2020 bike data.csv")


january = pd.read_csv(joining_january)
february = pd.read_csv(joining_february)
march = pd.read_csv(joining_march)
april = pd.read_csv(joining_april)
may = pd.read_csv(joining_may)
june = pd.read_csv(joining_june)
july = pd.read_csv(joining_july)
august = pd.read_csv(joining_august)
september = pd.read_csv(joining_september)
october = pd.read_csv(joining_october)
november = pd.read_csv(joining_november)
december = pd.read_csv(joining_december)
```
